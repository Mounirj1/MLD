## Data Analysis

### Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading data

```python
data = pd.read_csv('../input/breast-cancer-wisconsin.csv');
```

```python
data.head()
```

### Correlaton Matrix

```python
feature_list_1= list(data.columns[1:10])
plt.figure(figsize=(10,10))
sns.heatmap(data[feature_list_1].corr(), annot=True, square=True, cmap='coolwarm')
plt.show()
```

### Scatter Plot

```python
color_dic = {'M':'red', 'B':'green'}
colors = data['diagnosis'].map(lambda x: color_dic.get(x))
sm = pd.plotting.scatter_matrix(data[feature_list_1], c=colors, alpha=0.4, figsize=((15,15)));
plt.show()
```

### Box Plot

```python
plt.figure(figsize=(15,15))
for i, feature in enumerate(feature_list_1[1:]):
    rows = int(len(feature_list_1)/2)
    plt.subplot(rows, 2, i+1)
    sns.boxplot(x='diagnosis', y=feature, data=data, palette="Set2")

plt.tight_layout()
plt.show()
```
